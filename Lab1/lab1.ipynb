```python
import pandas as pd
import numpy as np

df_1 = pd.DataFrame() 
print('df_1\n',df_1)

data = {'color' : ['blue', 'green', 'yellow'],
        'object' : ['ball', 'pen', 'pencil'],
        'price' : [1.2, 1.0, 0.6]}
df_2 = pd.DataFrame(data)
print('df_2\n',df_2)

matrix = np.random.randint(10, size=(3, 5))
df_3 = pd.DataFrame(matrix)

print('df_3\n',df_3)

df_4 = pd.DataFrame(data, columns=['object', 'price'])
print('df_4\n',df_4)

# операції
print(df_2.head(2))
print(df_2.describe())
print(df_3.describe(include='all')) 

print(df_2.iloc[0]) 
print(df_2.iloc[[0, 2]])


print(df_2.loc[0]['color']) 
print(df_2.loc[df_2['price']>1]) 
print(df_3.loc[df_3[0]%2==0]) 

df_2_price = df_2.groupby(by=['price']).min()
print(df_2_price)

print(df_2.sort_values(by='price', ascending=False))

print(df_2['price'].isnull().head(3))
```

## Завдання 3 Matplotlib
створити набір даних лінійної функції з урахуванням помилок вимірювань, побудувати графік функції та згенерованих даних, обчислити похибку з використанням метрик MAE, MSE). Записати отримані результати у csv файл (формат: X,Y, Y_hat, mAE, mSE)

```python
import matplotlib.pyplot as plt
import numpy as np
import csv
from sklearn.metrics import mean_absolute_error, mean_squared_error

x = np.linspace(0, 10, 50)
y = [i**2 for i in x]

plt.title("Графік функції y = x^2") 
plt.xlabel("x") 
plt.ylabel("y") 
plt.grid()      
plt.plot(x, y) 

print(mean_absolute_error(x,y))
print(mean_squared_error(x,y))

FILENAME = "func.csv"
with open(FILENAME, "w", newline="") as file:
  writer = csv.writer(file)
  writer.writerow(["Name", "Value"])
  writer.writerows([
                    ["x",x],
                    ["y",y],
                    ["mAE", mean_absolute_error(x,y)],
                    ["mSE", mean_squared_error(x,y)],

  ])
```
